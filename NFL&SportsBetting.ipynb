```python
import pandas as pd
import requests
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read csv files

sports_gamers_watching_sports = Path('../project1_group6/Resources/% of sports gamblers who watch sports more than usual  when betting on a game.csv')
annual_legal_bets_superbowl = Path('../project1_group6/Resources/Annual Legal Bets on the Big Game.csv')
sports_betting_revenue_2018 = Path ('../project1_group6/Resources/Sports Betting Revenue since 2018.csv')
superbowl_ratings_viewership = Path('../project1_group6/Resources/Super Bowl Ratings and viewership, all-time.csv')
population_america_betting_on_sports = Path('../project1_group6/Resources/Population of America betting on sports.csv')

 
```

```python
sports_gamers_watching_sports = pd.read_csv(sports_gamers_watching_sports)
annual_legal_bets_superbowl = pd.read_csv(annual_legal_bets_superbowl)
sports_betting_revenue_2018 = pd.read_csv(sports_betting_revenue_2018)
superbowl_ratings_viewership = pd.read_csv(superbowl_ratings_viewership)
population_america_betting_on_sports = pd.read_csv(population_america_betting_on_sports) 
```

```python
population_america_betting_on_sports['Year'] = population_america_betting_on_sports['Year'].fillna('')
superbowl_ratings_viewership['Year'] = superbowl_ratings_viewership['Year'].fillna('')
sports_betting_revenue_2018['Year'] = sports_betting_revenue_2018['Year'].fillna('')
annual_legal_bets_superbowl['Year'] = annual_legal_bets_superbowl['Year'].fillna('')
```

```python
population_america_betting_on_sports['Year'] = population_america_betting_on_sports['Year'].astype(str)
superbowl_ratings_viewership['Year'] = superbowl_ratings_viewership['Year'].astype(str)
sports_betting_revenue_2018['Year'] = sports_betting_revenue_2018['Year'].astype(str)
annual_legal_bets_superbowl['Year'] = annual_legal_bets_superbowl['Year'].astype(str)
```

```python
mass_data = pd.merge(population_america_betting_on_sports, sports_betting_revenue_2018, on="Year", how="left")
mass_data = pd.merge(mass_data, annual_legal_bets_superbowl, on="Year", how="left")
mass_data = pd.merge(mass_data, superbowl_ratings_viewership, on="Year", how="left")
```

```python
mass_data['TV Vwrs'] = mass_data['TV Vwrs'].fillna('0')
```
