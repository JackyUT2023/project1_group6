```python
import pandas as pd
import requests
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read csv files

sports_gamers_watching_sports = Path('../project1_group6/Resources/% of sports gamblers who watch sports more than usual  when betting on a game.csv')
annual_legal_bets_superbowl = Path('../project1_group6/Resources/Annual Legal Bets on the Big Game.csv')
sports_betting_revenue_2018 = Path ('../project1_group6/Resources/Sports Betting Revenue since 2018.csv')
superbowl_ratings_viewership = Path('../project1_group6/Resources/Super Bowl Ratings and viewership, all-time.csv')
population_america_betting_on_sports = Path('../project1_group6/Resources/Population of America betting on sports.csv')
superbowl_viewership = Path('../project1_group6/Resources/SBtv - viewership with years.csv')

 
```

```python
sports_gamers_watching_sports = pd.read_csv(sports_gamers_watching_sports)
annual_legal_bets_superbowl = pd.read_csv(annual_legal_bets_superbowl)
sports_betting_revenue_2018 = pd.read_csv(sports_betting_revenue_2018)
superbowl_ratings_viewership = pd.read_csv(superbowl_ratings_viewership)
population_america_betting_on_sports = pd.read_csv(population_america_betting_on_sports) 
superbowl_viewership = pd.read_csv(superbowl_viewership)
```

```python
mass_data = pd.merge(population_america_betting_on_sports, sports_betting_revenue_2018, on="Year", how="inner")
mass_data = pd.merge(mass_data, annual_legal_bets_superbowl, on="Year", how="inner")
mass_data = pd.merge(mass_data, superbowl_ratings_viewership, on="Year", how="inner")
```

```python
mass_data_2013 = pd.merge(superbowl_ratings_viewership, annual_legal_bets_superbowl, on="Year", how="inner")
```

```python
mass_data['Year'] = mass_data['Year'].astype(int)
```

```python
years = mass_data_2013["Year"]
total_bets_millions = mass_data_2013["Total Bets (Millions $)"]
tv_viewers = mass_data_2013["Overall TV Vwrs"]

fig, ax1 = plt.subplots()

bar_width = 0.4
bar_positions = np.arange(len(years))
ax1.bar(bar_positions, total_bets_millions, width=bar_width, label='Total Bets (Millions $)')

ax2 = ax1.twinx()
ax2.plot(bar_positions, tv_viewers, color='orange', marker='o', label='TV Viewers (Millions)')

ax1.set_xlabel('Year')
ax1.set_xticks(bar_positions)
ax1.set_xticklabels(years)
ax1.set_ylabel('Total Bets (Millions $)')
ax2.set_ylabel('TV Viewers (Millions)')

lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left')

plt.title('Super Bowl: Total Bets (Millions $) and TV Viewers (Millions) by Year')

plt.savefig('../project1_group6/Resources/total_bets_tv_viewers.png')

plt.tight_layout()

plt.show()
```

```python
mass_data_pop_rev = pd.merge(sports_betting_revenue_2018, population_america_betting_on_sports, on="Year", how="inner")
```

```python
mass_data_pop_rev
```

```python
years = mass_data_pop_rev["Year"]
population_millions = mass_data_pop_rev["Population (Millions)"]
revenue = mass_data_pop_rev["Revenue"]

fig, ax1 = plt.subplots()

bar_width = 0.4
bar_positions1 = np.arange(len(years))
bar_positions2 = bar_positions1 + bar_width

ax1.bar(bar_positions1, population_millions, width=bar_width, label='Population (Millions)')

ax2 = ax1.twinx()
ax2.bar(bar_positions2, revenue, width=bar_width, color='orange', label='Revenue')

ax1.set_xlabel('Year')
ax1.set_xticks(bar_positions1 + bar_width / 2)
ax1.set_xticklabels(years)
ax1.set_ylabel('Population (Millions)')
ax2.set_ylabel('Revenue (Billions)')

lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left')

plt.title('Population (Millions) and Revenue by Year')

plt.savefig('../project1_group6/Resources/population_revenue.png')

plt.tight_layout()

plt.show()

```
