**Question 1: What is the most popular sports in United States?**
- Although there are many sports in United States, only 3 sports were taken into this analysis for comparison.

Limitation:
- we choose 3 sports which we heard most popular for this analysis only. 
- There might be some other sports also popular,like yoga, and jogging, but we don't have enough data for them. Maybe we just have not found them, or maybe there are no survey or study on them yet.

- The data sources as below:
    - https://www.espn.com/nfl/attendance
    - https://www.espn.com/mlb/attendance
    - https://www.espn.com/nba/attendance
- More explanation for the csv tables generated from about url.
    - The GMS, TOTAL, AVG, PCT columes are the home data;
    - The GMS.1, AVG.1, PCT.1 columes are the road data;
    - The GMS.2, AVG.2, PCT.2 columes are the overall data.

```python
# Dependencies
import pandas as pd
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
import json
import requests
from pprint import pprint
```

```python
# Read csv files

# NFL
epsn_nfl_attendance_2017_csv=Path("Resources/nfl_attendance_2017.csv",thousands=',')
epsn_nfl_attendance_2018_csv=Path("Resources/nfl_attendance_2018.csv",thousands=',')
epsn_nfl_attendance_2019_csv=Path("Resources/nfl_attendance_2019.csv",thousands=',')
epsn_nfl_attendance_2020_csv=Path("Resources/nfl_attendance_2020.csv",thousands=',')
epsn_nfl_attendance_2021_csv=Path("Resources/nfl_attendance_2021.csv",thousands=',')
epsn_nfl_attendance_2022_csv=Path("Resources/nfl_attendance_2022.csv",thousands=',')

# NBA
epsn_nba_attendance_2017_csv=Path("Resources/nba_attendance_2017.csv",thousands=',')
epsn_nba_attendance_2018_csv=Path("Resources/nba_attendance_2018.csv",thousands=',')
epsn_nba_attendance_2019_csv=Path("Resources/nba_attendance_2019.csv",thousands=',')
epsn_nba_attendance_2020_csv=Path("Resources/nba_attendance_2020.csv",thousands=',')
epsn_nba_attendance_2021_csv=Path("Resources/nba_attendance_2021.csv",thousands=',')
epsn_nba_attendance_2022_csv=Path("Resources/nba_attendance_2022.csv",thousands=',')

# MLB
epsn_mlb_attendance_2017_csv=Path("Resources/mlb_attendance_2017.csv",thousands=',')
epsn_mlb_attendance_2018_csv=Path("Resources/mlb_attendance_2018.csv",thousands=',')
epsn_mlb_attendance_2019_csv=Path("Resources/mlb_attendance_2019.csv",thousands=',')
    #epsn_mlb_attendance_2020_csv is not available due to pandemic closure.
epsn_mlb_attendance_2021_csv=Path("Resources/mlb_attendance_2021.csv",thousands=',')
epsn_mlb_attendance_2022_csv=Path("Resources/mlb_attendance_2022.csv",thousands=',')
```

```python
#NFL
epsn_nfl_attendance_2017=pd.read_csv(epsn_nfl_attendance_2017_csv)
NFL_2017_Ave_Attendance = round((epsn_nfl_attendance_2017["AVG.2"].str.replace(',', '').astype(int).mean()),0)
epsn_nfl_attendance_2018=pd.read_csv(epsn_nfl_attendance_2018_csv)
NFL_2018_Ave_Attendance = round((epsn_nfl_attendance_2018["AVG.2"].str.replace(',', '').astype(int).mean()),0)
epsn_nfl_attendance_2019=pd.read_csv(epsn_nfl_attendance_2019_csv)
NFL_2019_Ave_Attendance = round((epsn_nfl_attendance_2019["AVG.2"].str.replace(',', '').astype(int).mean()),0)
epsn_nfl_attendance_2020=pd.read_csv(epsn_nfl_attendance_2020_csv)
NFL_2020_Ave_Attendance = round((epsn_nfl_attendance_2020["AVG.2"].str.replace(',', '').astype(int).mean()),0)
epsn_nfl_attendance_2021=pd.read_csv(epsn_nfl_attendance_2021_csv)
NFL_2021_Ave_Attendance = round((epsn_nfl_attendance_2021["AVG.2"].str.replace(',', '').astype(int).mean()),0)
epsn_nfl_attendance_2022=pd.read_csv(epsn_nfl_attendance_2022_csv)
NFL_2022_Ave_Attendance = round((epsn_nfl_attendance_2022["AVG.2"].str.replace(',', '').astype(int).mean()),0)

#NBA
epsn_nba_attendance_2017=pd.read_csv(epsn_nba_attendance_2017_csv)
NBA_2017_Ave_Attendance = round((epsn_nba_attendance_2017["AVG.2"].str.replace(',', '').astype(int).mean()),0)
epsn_nba_attendance_2018=pd.read_csv(epsn_nba_attendance_2018_csv)
NBA_2018_Ave_Attendance = round((epsn_nba_attendance_2018["AVG.2"].str.replace(',', '').astype(int).mean()),0)
epsn_nba_attendance_2019=pd.read_csv(epsn_nba_attendance_2019_csv)
NBA_2019_Ave_Attendance = round((epsn_nba_attendance_2019["AVG.2"].str.replace(',', '').astype(int).mean()),0)
epsn_nba_attendance_2020=pd.read_csv(epsn_nba_attendance_2020_csv)
NBA_2020_Ave_Attendance = round((epsn_nba_attendance_2020["AVG.2"].str.replace(',', '').astype(int).mean()),0)
epsn_nba_attendance_2021=pd.read_csv(epsn_nba_attendance_2021_csv)
NBA_2021_Ave_Attendance = round((epsn_nba_attendance_2021["AVG.2"].str.replace(',', '').astype(int).mean()),0)
epsn_nba_attendance_2022=pd.read_csv(epsn_nba_attendance_2022_csv)
NBA_2022_Ave_Attendance = round((epsn_nba_attendance_2022["AVG.2"].str.replace(',', '').astype(int).mean()),0)

#MLB
epsn_mlb_attendance_2017=pd.read_csv(epsn_mlb_attendance_2017_csv)
MLB_2017_Ave_Attendance = round((epsn_mlb_attendance_2017["AVG.2"].str.replace(',', '').astype(int).mean()),0)
epsn_mlb_attendance_2018=pd.read_csv(epsn_mlb_attendance_2018_csv)
MLB_2018_Ave_Attendance = round((epsn_mlb_attendance_2018["AVG.2"].str.replace(',', '').astype(int).mean()),0)
epsn_mlb_attendance_2019=pd.read_csv(epsn_mlb_attendance_2019_csv)
MLB_2019_Ave_Attendance = round((epsn_mlb_attendance_2019["AVG.2"].str.replace(',', '').astype(int).mean()),0)
MLB_2020_Ave_Attendance = 0
epsn_mlb_attendance_2021=pd.read_csv(epsn_mlb_attendance_2021_csv)
MLB_2021_Ave_Attendance = round((epsn_mlb_attendance_2021["AVG.2"].str.replace(',', '').astype(int).mean()),0)
epsn_mlb_attendance_2022=pd.read_csv(epsn_mlb_attendance_2022_csv)
MLB_2022_Ave_Attendance = round((epsn_mlb_attendance_2022["AVG.2"].str.replace(',', '').astype(int).mean()),0)

## We could use a for loop structure to extract data instead of doing them one by one by hand. and that's something we can even impove it.

Attendance_data = {
    "Year":[2017,2018,2019,2020,2021,2022],
    "NFL":[NFL_2017_Ave_Attendance,NFL_2018_Ave_Attendance,NFL_2019_Ave_Attendance,NFL_2020_Ave_Attendance,NFL_2021_Ave_Attendance,NFL_2022_Ave_Attendance],
    "NBA":[NBA_2017_Ave_Attendance,NBA_2018_Ave_Attendance,NBA_2019_Ave_Attendance,NBA_2020_Ave_Attendance,NBA_2021_Ave_Attendance,NBA_2022_Ave_Attendance],
    "MLB":[MLB_2017_Ave_Attendance,MLB_2018_Ave_Attendance,MLB_2019_Ave_Attendance,MLB_2020_Ave_Attendance,MLB_2021_Ave_Attendance,MLB_2022_Ave_Attendance]
}

Attendance_data_df=pd.DataFrame(Attendance_data,columns=["Year","NFL","NBA","MLB"])
Attendance_data_df.set_index('Year')
```

```python
# data to plot
n_groups = len(Attendance_data_df["Year"])
NFL = Attendance_data_df["NFL"]
NBA = Attendance_data_df["NBA"]
MLB = Attendance_data_df["MLB"]

# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.2
opacity = 0.8

rects1 = plt.bar(index, NFL, bar_width,
alpha=opacity,
color='b',
label='NFL Attendence')

rects2 = plt.bar(index + bar_width, NBA, bar_width,
alpha=opacity,
color='g',
label='NBA Attendence')

rects2 = plt.bar(index + bar_width+ bar_width, MLB, bar_width,
alpha=opacity,
color='r',
label='MLB Attendence')

plt.xlabel('Year')
plt.ylabel('Average Attendance per Game')
plt.title('Attendance for Different Sports by Year')
plt.xticks(index + bar_width, Attendance_data_df["Year"])
plt.legend()

plt.tight_layout()
plt.savefig("Output/Attendance_for_different sports_by_year.png")
plt.show()
```
